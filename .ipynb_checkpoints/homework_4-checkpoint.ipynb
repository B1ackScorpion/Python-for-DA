### 1.1. Скачать следующие данные: kc-house-data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://gbcdn.mrgcdn.ru/uploads/asset/5349842/attachment/16ab1213a48b859e6e5e7407e10933ff.csv', sep=',')
df.head()
```

```python
import seaborn as sns
```

### 1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], bins=30)
plt.title('Распределение стоимости недвижимости')
plt.xlabel('euros')
plt.ylabel('Кол-во домов');
```

### 1.3. Изучите распределение квадратуры жилой


```python
plt.figure(figsize=(10, 5))
sns.histplot(df['sqft_living'], bins=30)
plt.title('Распределение квадратуры жилой')
plt.xlabel('euros')
plt.ylabel('Кол-во домов');
```

Вывод: у основной части домов жилая площадь занимает примерно между 1000 и 3000 метров квадратных. Самое большое число - между 1500 и 2500 метров квадратных.

### 1.4. Изучите распределение года постройки


```python
plt.figure(figsize=(10, 5))
sns.histplot(df['yr_built'], bins=30)
plt.title('Распределение годов постройки')
plt.xlabel('Год')
plt.ylabel('Кол-во домов');
```

Вывод: основная часть домов были построены примерно между 1950 и 2015 годами. Больше всего - в 2000 годах.

## Условие 2: Задача 2

### 2.1. Изучите распределение домов от наличия вида на набережную
- Постройте график
- Сделайте выводы

```python
df.loc[df['waterfront'] == 0, 'waterfront'] = 'Нет'
df.loc[df['waterfront'] == 1, 'waterfront'] = 'Есть'
data_w = df['waterfront'].value_counts()

names_w = data_w.index
values_w = data_w.values
```

```python
plt.pie(values_w, labels=names_w, autopct='%.1f%%', wedgeprops=dict(width=0.25));
```

Вывод: почти у всех домов отсутствует вид на набережную

### 2.2. Изучите распределение этажей домов

```python
plt.figure(figsize=(7, 5))
sns.histplot(df['floors'], bins=6)
plt.title('Распределение этажей домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во домов');
```

Вывод: в основном в домах 1 или 2 этажа. Самое большое кол-во домов с 1 этажем

### 2.3. Изучите распределение состояния домов


```python
data_c = df['condition'].value_counts()
names_c = data_c.index
values_c = data_c.values
```

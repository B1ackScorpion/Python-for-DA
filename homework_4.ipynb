## Условие 3: Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

Для проведения анализа характеристик недвижимости, влияющих на её стоимость, будем использовать набор данных и несколько визуализаций.

```python
df = pd.read_csv('kc-house-data.csv', sep=',')
df['price'] = df['price'] / 1000
```

```python
df.info()
```

Для начала уберем из датасета лишние данные. И для оставшихся построим матрицу корреляций, чтобы понять какие характеристики больше всего влияют на изменение цены. Так же для наглядности построим тепловую карту. Будет проводить анализ по 5 наиболее влиятельным характеристикам.

```python
data = df.drop(columns=['id', 'date', 'sqft_living15', 'sqft_lot15'])
data
```

```python
corr_matrix = data.corr().loc['price'].abs().sort_values(ascending=False)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3 ] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix.to_frame(), annot=True, linewidth=1, cmap='coolwarm');
```

С данной тепловой карты можно наглядно увидеть, что наиболее значемые характеристики для цены:
- жилая площадь
- оценка конструкии и дизайна
- жилая площадь, находящаяся выше земли
- кол-во ванных комнат в доме
- оценка открывающегося вида

Теперь проанализируем каждую характеристику отдельно с помощью графиков.

### Влияние жилой площади на стоимость

```python
plt.figure(figsize=(10, 5))

sns.scatterplot(x=data['sqft_living'], y=data['price'])

plt.title('График зависимости стоимости дома от жилой площади')
plt.xlabel('Жилая площадь (кв.м)')
plt.ylabel('Стоимость (1000 euros)');
```

Как видно из графика, существует положительная корреляция между площадью недвижимости и её стоимостью. С увеличением площади стоимость недвижимости также увеличивается. Линия регрессии на графике подтверждает эту линейную зависимость. Это логично, поскольку более просторные объекты недвижимости обычно стоят дороже.

### Влияние оценки конструкции и дизайна на стоимость

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=data['grade'], y=data['price'], showfliers=False);
plt.title('График зависимости стоимости дома от оценки качества конструкции и дизайна')
plt.xlabel('grade')
plt.ylabel('Стоимость (1000 euros)');
```

График показывает, что более высокие оценки конструкции и дизайна связаны с более высокой стоимостью недвижимости. Объекты с оценкой 13 имеют наибольшую стоимость, что указывает на значительное влияние высокого качества конструкции и дизайна на цену недвижимости.

### Влияние жилой площади, находящейся выше земли, на стоимость

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=data['sqft_above'], y=data['price'])
plt.title('График зависимости стоимости дома от площади дома выше земли')
plt.xlabel('площадь (м^2)')
plt.ylabel('Стоимость (1000 euros)');
```

Положительная корреляция между жилой площадью, находящейся выше земли, и стоимостью недвижимости аналогична общей жилой площади. Чем больше площадь выше земли, тем выше стоимость недвижимости. Это логично, поскольку большее количество жилой площади выше земли часто ассоциируется с большими и более просторными домами.

### Влияние количества ванных комнат на стоимость

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=data['bathrooms'], y=data['price']);
plt.title('График зависимости стоимости дома от кол-ва ванных комнат в доме')
plt.xlabel('кол-во')
plt.ylabel('Стоимость (1000 euros)')
plt.xticks(rotation=45);
```

График показывает, что в целом, стоимость недвижимости увеличивается с увеличением количества ванных комнат. Это согласуется с ожиданиями, так как дома с большим количеством ванных комнат обычно имеют большую площадь и удобства, что увеличивает их стоимость. На графике также видна значительная вариативность стоимости недвижимости для домов с более чем 5 ванными комнатами. Это может быть связано с тем, что в эту категорию попадают как роскошные особняки, так и дома с большим количеством комнат, но без дополнительных премиальных характеристик. Столбцы с высокими значениями ошибок указывают на высокую разбросанность данных, что может свидетельствовать о различиях в качестве и расположении недвижимости.

### Влияние оценки открывающегося вида на стоимость

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=data['view'], y=data['price'])
plt.title('График зависимости стоимости дома от оценки открывающегося вида')
plt.xlabel('оценка')
plt.ylabel('Стоимость (1000 euros)');
```

График показывает, что более высокие оценки открывающегося вида связаны с более высокой стоимостью недвижимости. Дома с оценкой вида 4 имеют наибольшую стоимость. Это указывает на то, что красивые виды значительно влияют на стоимость недвижимости, так как они добавляют привлекательности и уникальности объекту.
